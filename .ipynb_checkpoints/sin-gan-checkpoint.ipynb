```python
import torch
import torch.nn as nn
import torch.utils.data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import math
```

### Get the data

```python
def get_samples(num_of_points):
    train_data = torch.zeros((num_of_points, 2))
    x_signal = 2 * math.pi * torch.rand(num_of_points)
    y_signal = torch.sin(x_signal)
    
    train_data[:, 0] = x_signal
    train_data[:, 1] = y_signal
    
    labels = torch.zeros(num_of_points)
    
    train_set = [
        (train_data[i], labels[i]) for i in range(num_of_points)
    ]
    
    return train_set
```

```python
training_data = get_samples(100)

x_axis = [x[0] for x, _ in training_data]
y_axis = [x[1] for x, _ in training_data]

plt.plot(x_axis, y_axis, '.')
```

```python
print(type(training_data))
```

```python
batch_size = 50

train_loader = torch.utils.data.DataLoader(training_data, batch_size=batch_size, shuffle=True)
```

```python
for (samples, i) in enumerate(train_loader):
    print(samples, i)
```

### Define models

```python
class Discriminator(nn.Module):
    def __init__(self, input_size, output_size, layers, p=0.3):
        super().__init__()
        
       
        #layers
        
        layerlist = []
        n_in = input_size
        for i in layers:
            layerlist.append(nn.Linear(n_in, i))
            layerlist.append(nn.ReLU(inplace=True))
            layerlist.append(nn.Dropout(p))
            n_in = i
            
        layerlist.append(nn.Linear(layers[-1], output_size))
        layerlist.append(nn.Sigmoid())
        
        self.layers = nn.Sequential(*layerlist)
        
    def forward(self, x):
        return self.layers(x)  
```

```python
class Generator(nn.Module):
    def __init__(self, input_size, output_size, layers, p=0.3):
        super().__init__()
          
        layerlist = []
        n_in = input_size
        for i in layers:
            layerlist.append(nn.Linear(n_in, i))
            layerlist.append(nn.ReLU(inplace=True))
            layerlist.append(nn.Dropout(p))
            n_in = i
            
        layerlist.append(nn.Linear(layers[-1], output_size))
        
        
        self.layers = nn.Sequential(*layerlist)
        
        
    def forward(self, x):
        
        return torch.tanh(self.layers(x))
```

### Model hyperparameters

```python
input_size = 2

d_output_size = 1
g_output_size = 2

d_layers = [128, 64, 32]
g_layers = [32, 64]
```

```python
D = Discriminator(input_size, d_output_size, d_layers)
G = Generator(input_size, g_output_size, g_layers)
```

```python
print(D)
```

```python
print(G)
```

### Losses

```python
def loss_function(D_out, label):
    batch_size = D_out.size(0)
    
    if label == 1:
        labels = torch.ones(batch_size) * label #real labels = 1
    else:
        labels = torch.zeros(batch_size) #fake labels = 0
    
    criterion = nn.BCELoss()
    
    loss = criterion(D_out, labels)
    
    return loss
```

### Training

```python
import torch.optim as optim

# Optimizers
lr = 0.002

# Create optimizers for the discriminator and generator
d_optimizer = optim.Adam(D.parameters(), lr)
g_optimizer = optim.Adam(G.parameters(), lr)
```

```python
# training hyperparams
num_epochs = 100
# keep track of loss
losses = []

# train the network
D.train()
G.train()
for epoch in range(num_epochs):
    
    for batch_i, (real_samples, _) in enumerate(train_loader):
        print(f"Real samples: {real_samples}")
                 
        #   TRAIN THE DISCRIMINATOR
        
        d_optimizer.zero_grad()
        
        # 1. Train with real data

        # Compute the discriminator losses on real samples 
        D_real = D(real_samples)
        d_real_loss = loss_function(D_real, 1)
        
        # 2. Train with fake samples
        
        # Generate fake samples
        
        z = torch.randn((batch_size, 2))
        fake_samples = G(z)
        
        # Compute the discriminator losses on fake images        
        D_fake = D(fake_samples)
        d_fake_loss = loss_function(D_fake, 0)
        
        # add up loss and perform backprop
        d_loss = d_real_loss + d_fake_loss
        d_loss.backward()
        d_optimizer.step()
        
        
        #  TRAIN THE GENERATOR
        
        g_optimizer.zero_grad()
        
        # 1. Train with fake images and flipped labels
        
        # Generate fake samples
        z = torch.randn(batch_size, 2)
        fake_samples = G(z)
        
        
        # Compute the discriminator losses on fake images 
        D_fake = D(fake_samples)
        g_loss = loss_function(D_fake, 0) # use real loss to flip labels
        
        # perform backprop
        g_loss.backward()
        g_optimizer.step()


        if epoch % 2 == 0:
            print(f'Epoch {epoch+1} d_loss: {d_loss.item()},  g_loss: {g_loss.item()}')
    
    ## AFTER EACH EPOCH##
    # append discriminator loss and generator loss
    losses.append((d_loss.item(), g_loss.item()))
    
```

```python
fig, ax = plt.subplots()
losses = np.array(losses)
plt.plot(losses.T[0], label='Discriminator')
plt.plot(losses.T[1], label='Generator')
plt.title("Training Losses")
plt.legend()
```
